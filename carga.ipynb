# Leitura e "anonimização" da planilha de dados

É a primeira etapa do processo

Os arquivos de dados serão carregados a partir deste módulo em um novo dataset.

A medida que sofrerem ajustes/correções, os dadaset serão gravados para, em uma próxima execução, serem reutilizados.

```python
import pandas as pd
```

```python
def doFornecedorHide(Origem):
    # Cria a lista de nomes anônimos
    listaAnonima = []
    for i in range(Origem.size):
        if i in range(1, 10):
            Origem['Nome'][i] = 'Fornecedor 00' + str(i)
        elif i in range(10, 100):
            Origem['Nome'][i] = 'Fornecedor 0' + str(i)   
        else:
            Origem['Nome'][i] = 'Fornecedor' + str(i)
    return Origem
```

```python
def doCarga(diretorio, opcao = 2, arquivoXLS = 'Base_1milhão.xlsx'):
    """
    Abre o arquivo de dados de acordo com o solicitado em "Opções"
    0 - Carrega do zero - XLS
    1 - Carrega Anomizados - df_original
    2 - Carrega - df_originalAnonimo 
    3 - ETL_base_pronta_para_previsao
    4 - Subset de Pronta
    """ 
    if (opcao == 0):    # Carrega o Excel
        df_original = pd.read_excel(diretorio + arquivoXLS, sheet_name='Catalogo1')
        df_original = doFornecedorHide(df_original)     
        df_original.to_pickle(diretorio + "df_originalAnonimo.pkl")
    elif(opcao == 1):   # Carrega arquivo já com nomes de empresas tratados
        df_original = pd.read_pickle(diretorio + "df_original.pkl")
        df_original = doFornecedorHide(df_original)     
        df_original.to_pickle(diretorio + "df_originalAnonimo.pkl")
    elif(opcao == 2):   # Carrega arquivo já com nomes de empresas tratados
        df_original = pd.read_pickle(diretorio + "df_originalAnonimo.pkl")
    elif(opcao == 3):   # Carrega arquivo já totalmente tratado
        df_original = pd.read_pickle(diretorio + "ETL_base_pronta_para_previsao.pkl")
    else:               # Carrega arquivo já totalmente tratado
        df_original = pd.read_pickle(diretorio + "df_Mini.pkl")
        
    return df_original
```
